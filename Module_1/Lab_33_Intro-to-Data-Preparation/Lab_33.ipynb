```python
d['Profession'] = d['Profession'].str.lower()
```

```python
d['Profession'] = d['Profession'].ffill()
```

```python
d.loc[d['Profession'].str.startswith('etu'), 'Profession'] = 'student'
```

```python
d['BirthYear'] = d['BirthYear'].ffill()
```

```python
d[d['Amount'] > 30000]
# we can see that we have one major outlier
```

```python
#replace the outlier by median value
d.at[5, 'Amount'] = d['Amount'].median()
```

```python
#Truncate columns
d = d.astype({'BirthYear':'int'})
d = d.astype({'ClientID':'int'})
d = d.astype({'TransactionID':'int'})
```
