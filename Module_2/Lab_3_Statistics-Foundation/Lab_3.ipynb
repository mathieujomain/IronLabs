Challenge 4

The time between arrival of 60 patients at an intensive care unit were recorded to the nearest hour. The data are shown below. Time (hours) 0–19 20–39 40–59 60–79 80–99 100–119 120–139 140–159 160–179 Frequency 16 13 17 4 4 3 1 1 1 Determine the median, mean and standard deviation for this dataset.

mean: we need to calculate the mid point of each class
mid points: 9,5 - 29,5 - 49,5- 69,5 - 89,5 - 109,5 - 129,5 - 149,5 - 169,5
frequency * mid points: 152 - 383,5 - 841,5 - 278 - 358 - 328,5 - 129,5 - 149,5 - 169,5
mean = sum(frequency*mid point)/sum(frequency) = 46,5

median: L + W((N/2 - C) / F)
L = 40, W = 19, N = 60, C = 29, F = 17
median = 41,1

standard deviation: first weed need the variance using this formula: σ² = Σ(Fi * Mi^2) - (n * μ^2) / (n - 1)
μ represents the mean
Mi is the midpoint of the ith range of grouped data
Fi is the frequency of the ith range
n is the number of samples

Σ(Fi * Mi^2) = 209595
n = 60
μ2 = 46,5

variance = 1353,5
and std = sqrt(variance) = 36,8
