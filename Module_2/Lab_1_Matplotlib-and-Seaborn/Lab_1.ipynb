#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
fig, axes = plt.subplots(1, 2, figsize=(8,5))
      
axes[0].plot(x, x**2, label='y=x**2')
axes[0].plot(x,np.exp(x), label='y=ex')
axes[0].set_title("Normal scale")


axes[1].plot(x, x**2, label='y=x**2')
axes[1].plot(x, np.exp(x), label='y=ex')
axes[1].set_xscale("log")
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale")

plt.legend()
```

#### From the Fitbit data, we want to visually understand:

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
months = pd.DataFrame(fitbit['Steps'].groupby(fitbit['Months_encoded']).median())

months.reset_index(inplace=True)
months
```

```python
sns.lineplot(months['Months_encoded'], months['Steps'])
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
work_or_weekend = pd.DataFrame(fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median())

work_or_weekend.reset_index(inplace=True)
work_or_weekend
```

```python
#The data is separated between 0 and 1, so we need a plot bar to show the difference between both

work_or_weekend.plot.bar(x = "Work_or_Weekend", y = "Steps" )
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
filter = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

plt.style.use('seaborn-colorblind')


for i in filter:
    plt.scatter(fitbit['Steps'], fitbit[i], label=i)
    plt.xlabel('Steps')
    plt.ylabel(i)
    plt.legend()
    plt.show()
```
